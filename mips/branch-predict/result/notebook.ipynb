```python
import matplotlib.pyplot as plt
from matplotlib import rcParams

from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg')
```

```python
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = 'Lato'
rcParams['font.weight'] = 500
rcParams['font.style'] = 'normal'
rcParams['axes.labelweight'] = 500
rcParams['axes.labelcolor'] = 'gray'
rcParams['axes.titleweight'] = 500
```

```python
def split(s):
    return list(enumerate(map(float, s.strip().split('\n'))))

def read_data(filename):
    with open(filename, 'r') as fp:
        data = fp.read()
    return split(data)
```

```python
all_qsort = read_data('all-qsort.out')
btfnt_qsort = read_data('btfnt-qsort.out')
gshare_qsort = read_data('gshare-qsort.out')
lshare_qsort = read_data('lshare-qsort.out')
```

```python
all_calc = read_data('all-calc.out')
btfnt_calc = read_data('btfnt-calc.out')
gshare_calc = read_data('gshare-calc.out')
lshare_calc = read_data('lshare-calc.out')
```

```python
all_collatz = read_data('all-collatz.out')
btfnt_collatz = read_data('btfnt-collatz.out')
gshare_collatz = read_data('gshare-collatz.out')
lshare_collatz = read_data('lshare-collatz.out')
```

```python
def plot(name, data):
    ret = data[:200] + data[100::100]
    plt.plot(*zip(*ret), label=name, alpha=0.5)

def finish(name):
    plt.title(name.upper())
    plt.xlabel('number of branches')
    plt.ylabel('miss rate')
    plt.legend()
```

```python
plot('btfnt', btfnt_qsort)
plot('lshare', lshare_qsort)
plot('gshare', gshare_qsort)
plot('tournament', all_qsort)
finish('qsort')
```

```python
plot('btfnt', btfnt_calc)
plot('lshare', lshare_calc)
plot('gshare', gshare_calc)
plot('tournament', all_calc)
finish('calc')
```

```python
plot('btfnt', btfnt_collatz)
plot('lshare', lshare_collatz)
plot('gshare', gshare_collatz)
plot('tournament', all_collatz)
finish('collatz')
```
